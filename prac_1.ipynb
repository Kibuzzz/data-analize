```python
# Иванов павел
# Написать программу, которая вычисляет площадь фигуры,
# параметры которой подаются на вход. Фигуры, которые подаются на вход:
# треугольник, прямоугольник, круг. Результатом работы является словарь, где
# ключ – это название фигуры, а значение – это площадь.
import math

# переделать по трем сторонам
def calculate_triangle_area(base, height):
    return 0.5 * base * height

def calculate_rectangle_area(length, width):
    return length * width

def calculate_circle_area(radius):
    return math.pi * radius ** 2


print(calculate_triangle_area(base=5, height=10))
print(calculate_rectangle_area(length=4, width=6))
print(calculate_circle_area(radius=3))
```

```python
# Иванов павел
# Написать программу, которая на вход получает два числа и
# операцию, которую к ним нужно применить. Должны быть реализованы
# следующие операции: +, -, /, //, abs – модуль, pow или ** – возведение в
# степень.

operations = {
        '+': lambda x, y: x + y,
        '-': lambda x, y: x - y,
        '/': lambda x, y: x / y if y != 0 else "Error: Division by zero",
        '//': lambda x, y: x // y if y != 0 else "Error: Division by zero",
        'abs': lambda x, y: (abs(x), abs(y)),
        'pow': lambda x, y: pow(x, y),
        '**': lambda x, y: x ** y
    }


def calculate(a, b, operation):
	f = operations.get(operation, "Operation is not supported")
	return f(a, b)


print(calculate(1, 2, "+"))
print(calculate(1, 2, "-"))
print(calculate(1, 2, "/"))
print(calculate(1, 0, "/"))
print(calculate(1, 2, "//"))
print(calculate(1, 0, "//"))
print(calculate(1, 2, "abs"))
print(calculate(1, 2, "pow"))
```

```python
# Иванов павел
# Напишите программу, которая считыа сумма введённых чисел не будет равна 0 и
# после этого выводит суммувает с консоли числа (по
# одному в строке) до тех пор, пок квадратов всех считанных чисел.
def sum_till_zero():
    numbers = []  # Список для хранения введённых чисел
    total_sum = 0  # Переменная для отслеживания суммы чисел

    while True:
        number = int(input("Enter a number: "))  # Считываем число с консоли
        numbers.append(number)  # Добавляем число в список
        total_sum += number  # Обновляем текущую сумму
        print(f'Текущая сумма {total_sum}')

        if total_sum == 0:
            break

    sum_of_squares = sum(x ** 2 for x in numbers) 
    print("Sum of squares:", sum_of_squares)

# Запуск программы
sum_till_zero()
```

```python
# Иванов павел
# Напишите программу, которая выводит последовательность
# чисел, длинною N, где каждое число повторяется столько раз, чему оно равно.
# На вход программе передаётся неотрицательное целое число N. Например,
# если N = 7, то программа должна вывести 1 2 2 3 3 3 4. Вывод элементов списка
# через пробел – print(*list).

def print_nums(N):
    result = [] 
    current_number = 1 

    while len(result) < N:
        result.extend([current_number] * current_number)
        current_number += 1  

    result = result[:N]

    print(*result)

N = int(input("Введите число: "))
print_nums(N)
```

```python
# Иванов павел
# Даны два списка:
# А = [1,    2,   3,   4,   2,   1,   3,   4,   5,   6,   5,   4,   3,   2 ]
# В = [‘a’, ’b’, ’c’, ’c’, ’c’, ’b’, ’a’, ’c’, ’a’, ’a’, ’b’, ’c’, ’b’, ’a’]
# Создать словарь, в котором ключи – это содержимое списка В, а
# значения для ключей словаря – это сумма всех элементов списка А в
# соответствии с буквой, содержащийся на той же позиции в списке В.
# Пример результата программы: {‘a’ : 10, ‘b’ : 15, ‘c’ : 6}.
# Скачать и загрузить данные о стоимости домов в калифорнии,
# используя библиотеку sklearn.

A = [1, 2, 3, 4, 2, 1, 3, 4, 5, 6, 5, 4, 3, 2]
B = ['a', 'b', 'c', 'c', 'c', 'b', 'a', 'c', 'a', 'a', 'b', 'c', 'b', 'a']

result = {}

for i in range(len(A)):
    key = B[i]
    value = A[i]
    
    if key in result:
        result[key] += value
    else:
        result[key] = value

print(result)
```

```python
# Иванов Павел
# Узнать максимальное и минимальное значения медианной
# стоимости дома
from sklearn.datasets import fetch_california_housing
import pandas as pd

# Загрузка данных
data = fetch_california_housing(as_frame=True)

# Получение DataFrame
df = data['data']

# Добавление столбца с медианной стоимостью дома
df['median_house_value'] = data['target']
max_house_value = df['median_house_value'].max()
min_house_value = df['median_house_value'].min()

# Значения в тысячах доллатов
print(f"Максимальная медианная стоимость дома: {max_house_value}")
print(f"Минимальная медианная стоимость дома: {min_house_value}")
```

```python
# Иванов павел
# Используя метод apply(), вывести на экран название признака и его
# среднее значение.
from sklearn.datasets import fetch_california_housing

# Загрузка данных
data = fetch_california_housing(as_frame=True)

# Получение DataFrame
df = data['data']

# Вычисление среднего значения для каждого признака с использованием apply()
mean_value = df.apply(lambda x: x.mean())['MedInc']

print(f"Признак - MedInc, Среднее значение - {mean_value}")
```

```python
operations = {"read": "r", "write": "w", "execute": "x"}


def get_files():
	n = int(input()) # количество файлов
	files = {} 
	for _ in range(n):
		file, *rights = str(input()).split(" ")
		files[file] = rights
	return files


def get_input(files):
	m = int(input())
	for _ in range(m):
		operation, file = str(input()).split(" ")
		if operation not in operations:
			print("Error operation")
			continue
		if file not in files:
			print("File not found")
			continue
		if operations[operation] not in files[file]:
			print("ACESS DENIED")
			continue
		print("OK")


files = get_files()
get_input(files)
```
